```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

classes_1 = ['Mild', 'Moderate', 'Non', 'Very Mild']
precision_1 = [0.93, 0.71, 0.97, 0.93]
recall_1 = [0.87, 0.83, 0.97, 0.95]
f1_score_1 = [0.90, 0.77, 0.97, 0.94]

classes_2 = ['Mild', 'Moderate', 'Non', 'Very Mild']
precision_2 = [0.94, 0.67, 0.73, 0.72]
recall_2 = [0.24, 0.33, 0.96, 0.66]
f1_score_2 = [0.38, 0.44, 0.83, 0.69]

classes_3 = ['Mild', 'Moderate', 'Non', 'Very Mild']
precision_3 = [0.67, 0.33, 0.90, 0.92]
recall_3 = [0.90, 0.89, 0.88, 0.75]
f1_score_3 = [0.77, 0.48, 0.89, 0.83]

# Function to create a bar chart
def plot_classification_report(classes, precision, recall, f1_score, title):
    fig, ax = plt.subplots(figsize=(10, 6))

    width = 0.2  
    x = np.arange(len(classes))  

    # Creating bars for each metric
    rects1 = ax.bar(x - width, precision, width, label='Precision', color='royalblue')
    rects2 = ax.bar(x, recall, width, label='Recall', color='lightcoral')
    rects3 = ax.bar(x + width, f1_score, width, label='F1-score', color='darkorange')

    ax.set_xlabel('Class')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(classes)
    ax.legend()

    def add_values(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate(f'{height:.2f}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3), 
                        textcoords="offset points",
                        ha='center', va='bottom')

    add_values(rects1)
    add_values(rects2)
    add_values(rects3)

    # Display the plot
    plt.tight_layout()
    plt.show()

# Plotting graphs for each classification report
plot_classification_report(classes_1, precision_1, recall_1, f1_score_1, 'Classification Report for GA based ANN')
plot_classification_report(classes_2, precision_2, recall_2, f1_score_2, 'Classification Report for Random Forest (400 Trees)')
plot_classification_report(classes_3, precision_3, recall_3, f1_score_3, 'Classification Report for Fuzzy System')
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['GA based ANN', 'Random Forest (400 Trees)', 'Fuzzy System']
accuracy = [0.95, 0.74, 0.84]  

colors = ['#FF6347', '#1E90FF', '#32CD32'] 
def plot_accuracy(models, accuracy, title):
    # Set up the plot
    fig, ax = plt.subplots(figsize=(8, 5))

    bars = ax.bar(models, accuracy, color=colors)

    ax.set_xlabel('Model')
    ax.set_ylabel('Accuracy')
    ax.set_title(title)

    for i, acc in enumerate(accuracy):
        ax.text(i, acc + 0.01, f'{acc:.2f}', ha='center', va='bottom', fontsize=12, color='black')

    # Display the plot
    plt.tight_layout()
    plt.show()

plot_accuracy(models, accuracy, 'Accuracy Comparison for Different Models')
```
